```python
import pandas as pd
```

```python
train_df = pd.read_csv('train1.csv')

train_df
```

```python
train_label_df = pd.read_csv('trainLabels1.csv')



train_label_df
```

```python
train_df.isnull().sum().sum()
```

```python
from sklearn.model_selection import train_test_split

X=train_df
y = train_label_df

train_X,val_X,train_y,val_y=train_test_split(X,y,random_state = 0,train_size=0.8,test_size = 0.2)

train_X
```

```python
val_X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score

my_model = RandomForestClassifier(random_state = 0,n_estimators=200)

my_model.fit(X,y.values.ravel())

predictions=my_model.predict(val_X)

predictions

print(f'Mean : {accuracy_score(val_y,predictions)}')
```

```python
test_df = pd.read_csv('test1.csv')

test_df
```

```python
preds_test = my_model.predict(test_df)

output = pd.DataFrame({'Id': test_df.index + 1,
                       'Solution': preds_test})
output.to_csv('submission1.csv', index=False)
```
